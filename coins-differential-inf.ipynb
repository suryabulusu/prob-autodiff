```python
import torch
```

```python
def fixed_var(size, val):
    # like seen data -- no grads through this
    return torch.nn.functional.one_hot(torch.tensor(val), size).float()

heads, tails = fixed_var(2, 0), fixed_var(2, 1)
print("heads", heads)

def latent_var(size):
    # like unseen data / var / param -- grads can be pulled through this
    return torch.ones(size, requires_grad = True)

def coin():
    return latent_var(2)

def dice():
    return latent_var(6)

print("latent coin (all possibilities)", coin())
```

```python
def fair_coeff(size):
    return 1/size * torch.ones(size)

fair_coeff(2)
```

```python
# generating function for joint probability
# can use this for any generative story
# note that all Indicator func/vars occur linear [Darwiche 2000]

def gen_func(l_coin1, l_coin2):
    # simple case for coins
    # should sum to 1 -- we are enuemrating all possibilities effectively
    flip1 = fair_coeff(l_coin1.size()[0]) * l_coin1
    flip2 = fair_coeff(l_coin2.size()[0]) * l_coin2
    
    # einsum just to look cute
    return torch.einsum("ab, bc -> ac", flip1.view(-1, 1), flip2.view(1, -1)).sum() # (2,1) * (1,2)
```

```python
l_coin1, l_coin2 = coin(), coin()
gf = gen_func(l_coin1, l_coin2)

print("gf sums to 1 ->", gf)
```

```python
# to get marginal prob p(x1 = heads) .. x2 should take both possibilities, ie, latent_var for gf
l_coin2 = coin()
gf = gen_func(heads, l_coin2)

print(gf) # should be 1/2
```

```python
# to get joint prob value p(x1 = heads, x2 = tails) .. should just send these fixed_var to gf
gf = gen_func(heads, tails)
print(gf) # should be 1/4
```

```python
# to get constrained joint, ie, p(x1 = heads, x2) = [p(x1 = heads, x2 = heads), p(x1 = heads, x2 = tails)]
# we need a vector

# bad way:
gf1 = gen_func(heads, heads)
gf2 = gen_func(heads, tails)
print("bad way ->", [gf1, gf2])

# good way: USE DIFFERENTIATION!!
# when you do gf(x1 = heads, x2 = latent).backward() .. the grads in latent correspond to p(x1 = heads, x2 = heads)
# and p(x1 = heads, x2 = tails)
l_coin = coin()
gf = gen_func(heads, l_coin)
gf.backward()
print("good way ->", l_coin.grad)

# why is this better? its quicker, elegant.. what if your latent_vec is of size 100? :) 
```

```python
# once you have constrained joint and marginal -- conditional prob comes for free
# p(x2 | x1) = p(x2, x1 = e) / p(x1 = e)
# p(x2 | x1 = heads) = [p(x2 = heads | x1 = heads), p(x2 = tails | x1 = heads)]

l_coin = coin()
den = gen_func(heads, l_coin) # denominator = marginal
den.backward()
num = l_coin.grad # numerator = constrained joint

print(num / den)
```

```python
# a better way to get conditional -- log trick
# d/dx log(f(x)) = d/dx f(x) * 1/f(x)
# imagine if f(x) = gf(x1 = heads, x2 = latent) 
# numerator = d/dx f(x) -- which is constrained joint p(x1 = heads, x2)
# denominator = f(x) -- which is marginal p(x1 = heads)
# boom - we have the conditional by simply calling backward() on log of gf

l_coin = coin()
log_gf = gen_func(heads, l_coin).log()
log_gf.backward()

print(l_coin.grad)
```

```python
# slightly involved generative story -- to show how easy it is to compute probs for latents of diff sizes
weighted_die = 0.8 * torch.nn.functional.one_hot(torch.tensor(3), 6) + 0.2 * fair_coeff(6)
def gen_func(l_coin, l_die):
    flip = l_coin * fair_coeff(2)
    
    roll1 = l_die * fair_coeff(6)
    e1 = flip[1] * roll1 # if tails -- i consider fair die

    roll2 = l_die * weighted_die
    e2 = flip[0] * roll2 # if heads -- i consider weighted die
    return (e1 + e2).sum()

assert gen_func(coin(), dice()) == torch.tensor(1) # sum of all possibilities must be 1
```

```python
# p(x = heads, die = 4) -- joint
print(gen_func(heads, torch.nn.functional.one_hot(torch.tensor(4), 6)))

# p(x = tails, die = 3) -- joint
print(gen_func(tails, torch.nn.functional.one_hot(torch.tensor(3), 6)))
```

```python
# marginal p(x = heads)
print(gen_func(heads, dice()))
```

```python
# conditional p(die | x = heads) -- using log trick
l_die = dice()
log_gf = gen_func(heads, l_die).log()
log_gf.backward()
print(l_die.grad)
assert l_die.grad.sum() == torch.tensor(1)

# note the advantage -- (although minimal right now) -- is the minimal mental computation involved
# p(die | x = heads) = [p(die = i | x = heads) for i in 1..6]
# p(die = 1 | x = heads) = p(die = 1, x = heads) / p(x = heads)
# = 1/2 * 1/6 / 1/2 = 1/6
# p(die = 3 | x = heads) = (1/2 * (0.8 + 0.2/6)) / 1/2
# doing all this math takes slight time -- can be avoided with log().backward()
```

```python
# p(x = tails | die = 3)
l_coin = coin()
log_gf = gen_func(l_coin, torch.nn.functional.one_hot(torch.tensor(3), 6)).log()
log_gf.backward()
print(l_coin.grad)
```
